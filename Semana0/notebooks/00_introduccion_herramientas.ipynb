### Sintaxis Fundamental

Vamos a ver ejemplos b√°sicos que necesitar√°s en el curso:

````python
# 1. Variables y tipos de datos
# En Python, no necesitas declarar el tipo de variable

nombre = "Juan"              # String (texto)
edad = 25                     # Integer (entero)
altura = 1.75                 # Float (decimal)
es_estudiante = True          # Boolean (verdadero/falso)

print(f"Hola, me llamo {nombre}, tengo {edad} a√±os y mido {altura}m")
print(f"¬øSoy estudiante? {es_estudiante}")
````

````python
# 2. Listas (arrays)
# Las listas almacenan m√∫ltiples valores

calificaciones = [85, 90, 78, 92, 88]

print(f"Mis calificaciones: {calificaciones}")
print(f"Primera calificaci√≥n: {calificaciones[0]}")  # Los √≠ndices empiezan en 0
print(f"√öltima calificaci√≥n: {calificaciones[-1]}")  # -1 accede al √∫ltimo elemento
print(f"N√∫mero de calificaciones: {len(calificaciones)}")
````

````python
# 3. Operaciones matem√°ticas

a = 10
b = 3

print(f"Suma: {a + b}")
print(f"Resta: {a - b}")
print(f"Multiplicaci√≥n: {a * b}")
print(f"Divisi√≥n: {a / b}")
print(f"Divisi√≥n entera: {a // b}")
print(f"M√≥dulo (residuo): {a % b}")
print(f"Potencia: {a ** b}")
````

````python
# 4. Condicionales (if-else)

calificacion = 85

if calificacion >= 90:
    resultado = "Excelente"
elif calificacion >= 80:
    resultado = "Muy bien"
elif calificacion >= 70:
    resultado = "Bien"
else:
    resultado = "Necesitas mejorar"

print(f"Tu calificaci√≥n de {calificacion} es: {resultado}")
````

````python
# 5. Loops (ciclos)

# For loop - itera sobre una secuencia
print("Contando del 1 al 5:")
for i in range(1, 6):  # range(1, 6) genera n√∫meros del 1 al 5
    print(i)

print("\nRecorriendo una lista:")
frutas = ["manzana", "banana", "naranja"]
for fruta in frutas:
    print(f"Me gusta la {fruta}")
````

````python
# 6. Funciones
# Las funciones nos permiten reutilizar c√≥digo

def calcular_promedio(calificaciones):
    """
    Calcula el promedio de una lista de calificaciones.
    
    Parameters:
    -----------
    calificaciones : list
        Lista de n√∫meros
    
    Returns:
    --------
    float
        El promedio de las calificaciones
    """
    suma = sum(calificaciones)
    cantidad = len(calificaciones)
    promedio = suma / cantidad
    return promedio

# Usar la funci√≥n
mis_calificaciones = [85, 90, 78, 92, 88]
resultado = calcular_promedio(mis_calificaciones)
print(f"El promedio es: {resultado:.2f}")
````

````python
# 7. Librer√≠as (importar m√≥dulos)
# Las librer√≠as nos dan funcionalidad adicional

import math
import statistics

numeros = [10, 20, 30, 40, 50]

print(f"Promedio: {statistics.mean(numeros)}")
print(f"Mediana: {statistics.median(numeros)}")
print(f"Desviaci√≥n est√°ndar: {statistics.stdev(numeros):.2f}")
print(f"Ra√≠z cuadrada de 16: {math.sqrt(16)}")
print(f"Pi: {math.pi:.4f}")
````

## 4. GitHub - Control de Versiones üêô

### ¬øQu√© es GitHub?

**GitHub** es una plataforma de alojamiento de c√≥digo que usa **Git** (sistema de control de versiones) para:
- Guardar diferentes versiones de tu c√≥digo
- Colaborar con otros desarrolladores
- Compartir proyectos p√∫blicamente
- Hacer backup de tu trabajo

**Analog√≠a:** Piensa en GitHub como un "Google Drive" para c√≥digo, pero con superpoderes para rastrear cambios.

---

### Conceptos B√°sicos

#### 1. **Repositorio (Repo)**
- Carpeta que contiene tu proyecto
- Incluye c√≥digo, datos, documentaci√≥n
- Puede ser **p√∫blico** (todos lo ven) o **privado** (solo t√∫ y colaboradores)

**Ejemplo:** El repositorio de este curso:
```
https://github.com/HesusG/diagnostico-lineas-accion
```

#### 2. **Commit**
- "Foto" del estado de tu c√≥digo en un momento espec√≠fico
- Incluye mensaje describiendo qu√© cambiaste
- Permite volver a versiones anteriores si algo se rompe

**Ejemplo de commits:**
```
Commit 1: "A√±adir an√°lisis descriptivo"
Commit 2: "Corregir error en c√°lculo de media"
Commit 3: "Agregar visualizaciones con seaborn"
```

#### 3. **Branch (Rama)**
- Versi√≥n paralela del c√≥digo
- Te permite experimentar sin afectar la versi√≥n principal

**Ejemplo:**
- `main`: Versi√≥n estable del c√≥digo
- `feature-regresion`: Est√°s desarrollando nueva funcionalidad
- Cuando funcione, haces **merge** (fusionar) a `main`

#### 4. **Clone**
- Descargar una copia completa de un repositorio a tu computadora

#### 5. **Pull**
- Descargar los cambios m√°s recientes del repositorio remoto

#### 6. **Push**
- Subir tus cambios locales al repositorio remoto

- Valor num√©rico calculado de una muestra

- Patr√≥n de frecuencia de valores en datos
- Ejemplos: normal, uniforme, sesgada

- Medida de relaci√≥n lineal entre dos variables
- Rango: -1 (negativa perfecta) a +1 (positiva perfecta)
