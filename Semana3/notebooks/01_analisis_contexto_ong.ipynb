### 2.2 Contexto PESTEL

Analiza los factores del entorno que afectan a la ONG y la problem√°tica.

```python
import pandas as pd

# An√°lisis PESTEL
pestel = {
    'Pol√≠tico': [
        '[Ejemplo: Pol√≠ticas p√∫blicas de asistencia social]',
        '[Leyes o regulaciones relevantes]'
    ],
    'Econ√≥mico': [
        '[Ejemplo: Nivel de pobreza en la zona]',
        '[Disponibilidad de financiamiento]'
    ],
    'Social': [
        '[Ejemplo: Percepci√≥n de la comunidad sobre el problema]',
        '[Niveles educativos, estructura familiar]'
    ],
    'Tecnol√≥gico': [
        '[Ejemplo: Acceso a internet/tecnolog√≠a]',
        '[Herramientas digitales disponibles]'
    ],
    'Ecol√≥gico': [
        '[Ejemplo: Factores ambientales que afectan]',
        '[Sostenibilidad de recursos]'
    ],
    'Legal': [
        '[Ejemplo: Marco legal de ONGs]',
        '[Requisitos regulatorios]'
    ]
}

print("AN√ÅLISIS PESTEL")
print("="*60)
for dimension, factores in pestel.items():
    print(f"\n{dimension.upper()}:")
    for i, factor in enumerate(factores, 1):
        print(f"  {i}. {factor}")
```

## 3. Alineaci√≥n con ODS y Derechos Humanos

```python
# ODS a los que contribuye la ONG
ods_contribucion = {
    'ODS Principal': {
        'numero': 0,  # Ej: 16 para Paz, Justicia e Instituciones S√≥lidas
        'nombre': '',  # Ej: "Paz, Justicia e Instituciones S√≥lidas"
        'justificacion': """
        [Explica c√≥mo la ONG contribuye a este ODS.
        Ejemplo: "Contribuye al ODS 16 al fortalecer instituciones locales..."]  
        """
    },
    'ODS Secundarios': [
        {
            'numero': 0,
            'nombre': '',
            'justificacion': ''
        },
        # Agrega m√°s si aplica
    ]
}

print("ALINEACI√ìN CON ODS")
print("="*60)
print(f"\nODS PRINCIPAL: ODS {ods_contribucion['ODS Principal']['numero']}")
print(f"{ods_contribucion['ODS Principal']['nombre']}")
print(f"\nJustificaci√≥n:")
print(ods_contribucion['ODS Principal']['justificacion'].strip())

if ods_contribucion['ODS Secundarios']:
    print("\n" + "="*60)
    print("ODS SECUNDARIOS:")
    for ods in ods_contribucion['ODS Secundarios']:
        if ods['numero'] > 0:
            print(f"\n- ODS {ods['numero']}: {ods['nombre']}")
            print(f"  {ods['justificacion']}")
```

## 4. Stakeholders y Ecosistema

### 4.1 Mapa de Stakeholders

Identifica los actores clave que interact√∫an con la ONG.

```python
stakeholders = {
    'Beneficiarios Directos': {
        'descripcion': '',  # Qui√©nes son
        'cantidad': 0,  # N√∫mero aproximado
        'caracteristicas': []  # Lista de caracter√≠sticas
    },
    'Beneficiarios Indirectos': {
        'descripcion': '',  # Ej: "Familias de los ni√±os atendidos"
        'cantidad': 0,
        'como_impacta': ''  # C√≥mo les beneficia indirectamente
    },
    'Donantes/Financiadores': [
        {' nombre': '', 'tipo': '', 'aporte': ''}  # Ej: tipo=Gobierno, Empresa, Fundaci√≥n
    ],
    'Aliados Estrat√©gicos': [
        {'nombre': '', 'tipo_alianza': ''}  # Otras ONGs, universidades, etc.
    ],
    'Gobierno': {
        'nivel': '',  # Municipal, Estatal, Federal
        'tipo_relacion': ''  # Colaboraci√≥n, regulaci√≥n, etc.
    }
}

print("MAPA DE STAKEHOLDERS")
print("="*60)
print(f"\nBENEFICIARIOS DIRECTOS: ~{stakeholders['Beneficiarios Directos']['cantidad']} personas")
print(f"Descripci√≥n: {stakeholders['Beneficiarios Directos']['descripcion']}")
print(f"\nBENEFICIARIOS INDIRECTOS: ~{stakeholders['Beneficiarios Indirectos']['cantidad']} personas")
print(f"Impacto: {stakeholders['Beneficiarios Indirectos']['como_impacta']}")

print("\n" + "="*60)
print("FINANCIADORES:")
for don in stakeholders['Donantes/Financiadores']:
    if don.get('nombre'):
        print(f"- {don['nombre']} ({don['tipo']}): {don['aporte']}")

print("\n" + "="*60)
print("ALIADOS:")
for aliado in stakeholders['Aliados Estrat√©gicos']:
    if aliado.get('nombre'):
        print(f"- {aliado['nombre']}: {aliado['tipo_alianza']}")
```

### 4.2 Visualizaci√≥n de Stakeholders

Crea un diagrama simple de stakeholders usando matplotlib.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Crear figura
fig, ax = plt.subplots(figsize=(10, 8))

# ONG en el centro
ong_circle = plt.Circle((0.5, 0.5), 0.08, color='#0062A4', alpha=0.8)
ax.add_patch(ong_circle)
ax.text(0.5, 0.5, 'TU\nONG', ha='center', va='center', 
        fontsize=12, fontweight='bold', color='white')

# Stakeholders alrededor (puedes ajustar posiciones)
stakeholder_positions = {
    'Beneficiarios': (0.5, 0.85),
    'Donantes': (0.15, 0.7),
    'Gobierno': (0.15, 0.3),
    'Aliados': (0.85, 0.7),
    'Comunidad': (0.85, 0.3),
    'Voluntarios': (0.5, 0.15)
}

colors = {'Beneficiarios': '#8CC63F', 'Donantes': '#FF6F31', 
          'Gobierno': '#939598', 'Aliados': '#009FDA',
          'Comunidad': '#D1D3D4', 'Voluntarios': '#003E7E'}

for stakeholder, (x, y) in stakeholder_positions.items():
    circle = plt.Circle((x, y), 0.06, color=colors.get(stakeholder, '#D1D3D4'), alpha=0.7)
    ax.add_patch(circle)
    ax.text(x, y, stakeholder, ha='center', va='center', fontsize=9, fontweight='bold')
    
    # L√≠nea conectando con ONG
    ax.plot([0.5, x], [0.5, y], 'k--', alpha=0.3, linewidth=1)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')
plt.title('Ecosistema de Stakeholders de la ONG', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

print("\nüí° TIP: Personaliza este diagrama agregando los stakeholders espec√≠ficos de tu ONG")
```

## 5. Recursos y Capacidades

### 5.1 Recursos Humanos

```python
recursos_humanos = {
    'empleados_tiempo_completo': 0,
    'empleados_medio_tiempo': 0,
    'voluntarios_regulares': 0,
    'voluntarios_ocasionales': 0,
    'consejo_directivo': 0,
    'areas_clave': []  # Ej: ['Administraci√≥n', 'Operaciones', 'Fundraising']
}

total_personal = (recursos_humanos['empleados_tiempo_completo'] + 
                 recursos_humanos['empleados_medio_tiempo'])
total_voluntarios = (recursos_humanos['voluntarios_regulares'] + 
                    recursos_humanos['voluntarios_ocasionales'])

print("RECURSOS HUMANOS")
print("="*60)
print(f"Personal remunerado: {total_personal}")
print(f"  - Tiempo completo: {recursos_humanos['empleados_tiempo_completo']}")
print(f"  - Medio tiempo: {recursos_humanos['empleados_medio_tiempo']}")
print(f"\nVoluntarios: {total_voluntarios}")
print(f"  - Regulares: {recursos_humanos['voluntarios_regulares']}")
print(f"  - Ocasionales: {recursos_humanos['voluntarios_ocasionales']}")
print(f"\n√Åreas clave:")
for area in recursos_humanos['areas_clave']:
    print(f"  - {area}")
```

### 5.2 Recursos Financieros

```python
# Presupuesto anual aproximado (si te lo comparten)
presupuesto = {
    'anual_total': 0,  # En pesos mexicanos
    'fuentes_ingreso': {
        'Donaciones individuales': 0,  # % del total
        'Donaciones corporativas': 0,
        'Fondos gubernamentales': 0,
        'Fundaciones': 0,
        'Eventos/Fundraising': 0,
        'Otros': 0
    },
    'principales_gastos': []  # Ej: ['Personal (60%)', 'Operaciones (30%)', 'Admin (10%)']
}

print("RECURSOS FINANCIEROS")
print("="*60)
if presupuesto['anual_total'] > 0:
    print(f"Presupuesto anual: ${presupuesto['anual_total']:,.0f} MXN")
    print(f"\nFuentes de ingreso:")
    for fuente, porcentaje in presupuesto['fuentes_ingreso'].items():
        if porcentaje > 0:
            print(f"  - {fuente}: {porcentaje}%")
else:
    print("Informaci√≥n financiera no disponible o confidencial")

if presupuesto['principales_gastos']:
    print(f"\nDistribuci√≥n de gastos:")
    for gasto in presupuesto['principales_gastos']:
        print(f"  - {gasto}")
```
