## 2. ValidaciÃ³n de KPIs Identificados

Crea un dashboard ejecutivo con los KPIs principales.

```python
# Define tus KPIs (basado en Semana 3)
kpis_finales = pd.DataFrame([
    {
        'KPI': 'SatisfacciÃ³n del Beneficiario',
        'Valor_Actual': 7.5,
        'Meta': 8.5,
        'Cumplimiento_%': 88.2,
        'Prioridad': 'Alta',
        'Frecuencia_Medicion': 'Mensual'
    },
    {
        'KPI': 'Tiempo de AtenciÃ³n Promedio',
        'Valor_Actual': 25.3,
        'Meta': 20.0,
        'Cumplimiento_%': 79.1,
        'Prioridad': 'Media',
        'Frecuencia_Medicion': 'Semanal'
    },
    {
        'KPI': '% Beneficiarios Satisfechos',
        'Valor_Actual': 68.5,
        'Meta': 80.0,
        'Cumplimiento_%': 85.6,
        'Prioridad': 'Alta',
        'Frecuencia_Medicion': 'Mensual'
    },
    # Agrega mÃ¡s KPIs segÃºn tu anÃ¡lisis
])

print("TABLA DE KPIs FINALES")
print("="*70)
print(kpis_finales.to_string(index=False))

# VisualizaciÃ³n de cumplimiento de KPIs
fig, ax = plt.subplots(figsize=(12, 6))

colors = ['#8CC63F' if x >= 90 else '#FF6F31' if x < 80 else '#939598' 
          for x in kpis_finales['Cumplimiento_%']]

bars = ax.barh(kpis_finales['KPI'], kpis_finales['Cumplimiento_%'], color=colors)
ax.axvline(90, color='green', linestyle='--', linewidth=2, alpha=0.7, label='Ã“ptimo: 90%')
ax.axvline(80, color='orange', linestyle='--', linewidth=2, alpha=0.7, label='Aceptable: 80%')

ax.set_title('Cumplimiento de KPIs EstratÃ©gicos', fontsize=16, fontweight='bold')
ax.set_xlabel('% de Cumplimiento', fontsize=12)
ax.set_xlim(0, 100)
ax.legend()
ax.grid(True, alpha=0.3, axis='x')

for i, bar in enumerate(bars):
    width = bar.get_width()
    ax.text(width + 2, bar.get_y() + bar.get_height()/2,
            f'{width:.1f}%', ha='left', va='center', fontweight='bold')

plt.tight_layout()
plt.savefig('cumplimiento_kpis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nâœ… GrÃ¡fico guardado: cumplimiento_kpis.png")
```
