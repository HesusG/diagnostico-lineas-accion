## 1. PriorizaciÃ³n de Problemas Identificados

Usa una matriz de priorizaciÃ³n para decidir quÃ© abordar primero.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Matriz de priorizaciÃ³n: Impacto vs Esfuerzo
problemas = pd.DataFrame([
    {'Problema': 'Tiempo de espera largo', 'Impacto': 8, 'Esfuerzo': 4},
    {'Problema': 'Baja satisfacciÃ³n en TrÃ¡mites', 'Impacto': 9, 'Esfuerzo': 6},
    {'Problema': 'Falta de sistema de quejas', 'Impacto': 7, 'Esfuerzo': 2},
    {'Problema': 'Datos no digitalizados', 'Impacto': 6, 'Esfuerzo': 5},
    {'Problema': 'Horarios limitados', 'Impacto': 5, 'Esfuerzo': 7},
    # Agrega problemas identificados en tu diagnÃ³stico
])

# ClasificaciÃ³n
def clasificar_prioridad(row):
    if row['Impacto'] >= 7 and row['Esfuerzo'] <= 4:
        return 'Quick Win'
    elif row['Impacto'] >= 7 and row['Esfuerzo'] > 4:
        return 'Proyecto Mayor'
    elif row['Impacto'] < 7 and row['Esfuerzo'] <= 4:
        return 'Mejora Menor'
    else:
        return 'Evitar'

problemas['Prioridad'] = problemas.apply(clasificar_prioridad, axis=1)

print("MATRIZ DE PRIORIZACIÃ“N DE PROBLEMAS")
print("="*70)
print(problemas.to_string(index=False))

# VisualizaciÃ³n
fig, ax = plt.subplots(figsize=(12, 8))

colors = {'Quick Win': '#8CC63F', 'Proyecto Mayor': '#0062A4', 
          'Mejora Menor': '#939598', 'Evitar': '#FF6F31'}

for prioridad, color in colors.items():
    subset = problemas[problemas['Prioridad'] == prioridad]
    ax.scatter(subset['Esfuerzo'], subset['Impacto'], 
              c=color, s=300, alpha=0.6, edgecolors='black', linewidth=2,
              label=prioridad)
    
    for idx, row in subset.iterrows():
        ax.annotate(row['Problema'], (row['Esfuerzo'], row['Impacto']),
                   fontsize=9, ha='center', va='center', fontweight='bold')

# LÃ­neas divisorias
ax.axhline(7, color='black', linestyle='--', alpha=0.3)
ax.axvline(4, color='black', linestyle='--', alpha=0.3)

# Etiquetas de cuadrantes
ax.text(2, 8.5, 'QUICK WINS', fontsize=12, fontweight='bold', 
        ha='center', color='#8CC63F', alpha=0.7)
ax.text(7, 8.5, 'PROYECTOS MAYORES', fontsize=12, fontweight='bold',
        ha='center', color='#0062A4', alpha=0.7)
ax.text(2, 5.5, 'MEJORAS MENORES', fontsize=12, fontweight='bold',
        ha='center', color='#939598', alpha=0.7)
ax.text(7, 5.5, 'EVITAR', fontsize=12, fontweight='bold',
        ha='center', color='#FF6F31', alpha=0.7)

ax.set_xlabel('Esfuerzo Requerido (1=Bajo, 10=Alto)', fontsize=12, fontweight='bold')
ax.set_ylabel('Impacto Potencial (1=Bajo, 10=Alto)', fontsize=12, fontweight='bold')
ax.set_title('Matriz de PriorizaciÃ³n: Impacto vs Esfuerzo', 
            fontsize=14, fontweight='bold', pad=20)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.legend(loc='upper left', fontsize=10)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('matriz_priorizacion.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nâœ… GrÃ¡fico guardado: matriz_priorizacion.png")
```

## 3. Plan de ImplementaciÃ³n con Timeline

```python
# Gantt chart simplificado
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define fechas de inicio y duraciÃ³n
inicio = datetime(2024, 7, 1)

plan_implementacion = pd.DataFrame([
    {'Linea': 'BuzÃ³n de quejas', 'Inicio': 0, 'Duracion': 30, 'Color': '#8CC63F'},
    {'Linea': 'Sistema de turnos', 'Inicio': 15, 'Duracion': 90, 'Color': '#0062A4'},
    {'Linea': 'CapacitaciÃ³n personal', 'Inicio': 30, 'Duracion': 60, 'Color': '#FF6F31'},
    {'Linea': 'Mejora infraestructura', 'Inicio': 60, 'Duracion': 120, 'Color': '#939598'},
])

fig, ax = plt.subplots(figsize=(14, 6))

for idx, row in plan_implementacion.iterrows():
    ax.barh(row['Linea'], row['Duracion'], left=row['Inicio'],
            height=0.5, color=row['Color'], alpha=0.8, edgecolor='black')
    
    # Etiqueta de duraciÃ³n
    ax.text(row['Inicio'] + row['Duracion']/2, idx, 
            f"{row['Duracion']} dÃ­as",
            ha='center', va='center', fontweight='bold', color='white', fontsize=10)

ax.set_xlabel('DÃ­as desde inicio del proyecto', fontsize=12, fontweight='bold')
ax.set_title('Cronograma de ImplementaciÃ³n de LÃ­neas de AcciÃ³n',
            fontsize=14, fontweight='bold', pad=20)
ax.set_xlim(0, 180)
ax.grid(True, alpha=0.3, axis='x')
ax.invert_yaxis()

# Marcar hitos
ax.axvline(30, color='red', linestyle='--', linewidth=2, alpha=0.5, label='Mes 1')
ax.axvline(90, color='orange', linestyle='--', linewidth=2, alpha=0.5, label='Mes 3')
ax.legend()

plt.tight_layout()
plt.savefig('cronograma_implementacion.png', dpi=300, bbox_inches='tight')
plt.show()

print("âœ… Cronograma guardado: cronograma_implementacion.png")
```

## 4. Recursos Necesarios y Presupuesto

```python
# Presupuesto consolidado
presupuesto_total = pd.DataFrame([
    {'Linea_Accion': 'Sistema de turnos digitales', 'Tecnologia': 0, 'Personal': 3000, 'Infraestructura': 2000, 'Otros': 0},
    {'Linea_Accion': 'BuzÃ³n de quejas', 'Tecnologia': 500, 'Personal': 1000, 'Infraestructura': 500, 'Otros': 0},
    {'Linea_Accion': 'CapacitaciÃ³n personal', 'Tecnologia': 2000, 'Personal': 5000, 'Infraestructura': 0, 'Otros': 1000},
    # Agrega segÃºn tus lÃ­neas de acciÃ³n
])

presupuesto_total['Total'] = presupuesto_total[['Tecnologia', 'Personal', 'Infraestructura', 'Otros']].sum(axis=1)

print("PRESUPUESTO CONSOLIDADO DE LÃNEAS DE ACCIÃ“N")
print("="*70)
print(presupuesto_total.to_string(index=False))
print(f"\nTOTAL GENERAL: ${presupuesto_total['Total'].sum():,.2f} MXN")

# GrÃ¡fico de distribuciÃ³n
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Por categorÃ­a
categorias = ['Tecnologia', 'Personal', 'Infraestructura', 'Otros']
totales_cat = [presupuesto_total[cat].sum() for cat in categorias]
colors_cat = ['#0062A4', '#8CC63F', '#FF6F31', '#939598']

ax1.pie(totales_cat, labels=categorias, autopct='%1.1f%%', colors=colors_cat,
        startangle=90, textprops={'fontsize': 11, 'fontweight': 'bold'})
ax1.set_title('DistribuciÃ³n del Presupuesto por CategorÃ­a', 
             fontsize=14, fontweight='bold')

# Por lÃ­nea de acciÃ³n
presupuesto_total.plot(x='Linea_Accion', y='Total', kind='barh', ax=ax2,
                       color='#0062A4', legend=False)
ax2.set_title('Presupuesto por LÃ­nea de AcciÃ³n', fontsize=14, fontweight='bold')
ax2.set_xlabel('Presupuesto (MXN)', fontsize=12)
ax2.set_ylabel('')
ax2.grid(True, alpha=0.3, axis='x')

for i, total in enumerate(presupuesto_total['Total']):
    ax2.text(total + 200, i, f'${total:,.0f}', va='center', fontweight='bold')

plt.tight_layout()
plt.savefig('presupuesto_lineas_accion.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nâœ… GrÃ¡fico guardado: presupuesto_lineas_accion.png")
```
