## Parte 3: Pruebas de Hip√≥tesis

### 3.1 Prueba t de Una Muestra

**üìù SOLUCI√ìN 3.1a - Hip√≥tesis:**

- **H‚ÇÄ (Nula):** Œº_confianza_financiera = 7.0 (la confianza financiera promedio es igual al umbral aceptable)

- **H‚ÇÅ (Alternativa):** Œº_confianza_financiera < 7.0 (la confianza financiera promedio es menor que el umbral)

**Tipo de prueba:** Unilateral izquierda (one-tailed test, left-sided)

**Nivel de significancia:** Œ± = 0.05

```python
# Paso 2: Exploraci√≥n Descriptiva
variable = df['confianza_financiera']

media_obs = variable.mean()
desv_est = variable.std()
n = len(variable)

print("üìä ESTAD√çSTICAS DESCRIPTIVAS: confianza_financiera\n")
print(f"Media observada: {media_obs:.3f}")
print(f"Desviaci√≥n est√°ndar: {desv_est:.3f}")
print(f"Tama√±o de muestra: {n}")
print(f"Valor de referencia: 7.0")
print(f"Diferencia: {media_obs - 7.0:.3f}")
```

```python
# Paso 3: Prueba t de Una Muestra
t_stat, p_value = stats.ttest_1samp(variable, popmean=7.0, alternative='less')

print("\n" + "="*70)
print("üß™ RESULTADOS DE LA PRUEBA T DE UNA MUESTRA\n")
print(f"Estad√≠stico t: {t_stat:.4f}")
print(f"Valor p (unilateral): {p_value:.4f}")
print(f"Nivel de significancia Œ±: 0.05")
print("="*70)
```

```python
# Paso 4: Decisi√≥n
alpha = 0.05

if p_value < alpha:
    decision = "Rechazamos H‚ÇÄ"
    conclusion = "Hay evidencia estad√≠stica suficiente para afirmar que la confianza financiera promedio es MENOR que 7.0"
else:
    decision = "No rechazamos H‚ÇÄ"
    conclusion = "No hay evidencia estad√≠stica suficiente para afirmar que la confianza financiera promedio es menor que 7.0"

print("\nüìã DECISI√ìN ESTAD√çSTICA:\n")
print(f"Decisi√≥n: {decision}")
print(f"\nConclusi√≥n: {conclusion}")
```

**üìù SOLUCI√ìN 3.1b - Interpretaci√≥n:**

a) **Interpretaci√≥n en contexto:**

**Escenario 1 (Si rechazamos H‚ÇÄ):**
La confianza financiera promedio est√° **estad√≠sticamente por debajo** del umbral de 7.0. Esto significa que los donadores perciben el manejo financiero de la Fundaci√≥n como **insuficiente**. Este es un hallazgo **cr√≠tico** porque la confianza financiera es un predictor clave de retenci√≥n de donadores seg√∫n la literatura (Sargeant, 2001).

**Escenario 2 (Si no rechazamos H‚ÇÄ):**
No hay evidencia de que la confianza est√© por debajo de 7.0. Sin embargo, si la media observada est√° cerca de 7.0 (ej: 7.1-7.3), aunque no sea estad√≠sticamente menor, sigue siendo un √°rea de **vigilancia** pues no hay mucho margen de seguridad.

b) **Recomendaciones:**

1. **Transparencia financiera:** Publicar reportes financieros detallados y accesibles en el sitio web
2. **Comunicaci√≥n de impacto:** Mostrar claramente c√≥mo se utilizan los fondos (ej: "Por cada $100 donados, $85 van directamente a programas")
3. **Auditor√≠as externas:** Obtener certificaciones de transparencia (ej: Conf√≠a, GuideStar)
4. **Storytelling con n√∫meros:** Comunicar historias de impacto con cifras concretas
5. **Seguimiento trimestral:** Medir esta variable cada 3 meses para detectar cambios tempranos

### 3.2 Prueba t de Dos Muestras Independientes: Voluntariado

**üìù SOLUCI√ìN 3.2a - Hip√≥tesis:**

- **H‚ÇÄ (Nula):** Œº_voluntarios = Œº_no_voluntarios (no hay diferencia en conexi√≥n emocional)

- **H‚ÇÅ (Alternativa):** Œº_voluntarios > Œº_no_voluntarios (voluntarios tienen mayor conexi√≥n emocional)

**Tipo de prueba:** Unilateral derecha (one-tailed test, right-sided)

**Nivel de significancia:** Œ± = 0.05

```python
# Paso 2: Exploraci√≥n Descriptiva
print("üìä ESTAD√çSTICAS DESCRIPTIVAS POR GRUPO\n")
print("Variable: conexion_emocional (escala 1-10)")
print("Grupos: Voluntariado (S√≠ / No)\n")

stats_by_group = df.groupby('voluntariado')['conexion_emocional'].agg(['mean', 'std', 'count'])
stats_by_group.columns = ['Media', 'Desv. Est√°ndar', 'n']
print(stats_by_group)

# Visualizaci√≥n
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='voluntariado', y='conexion_emocional', palette='Set2')
sns.stripplot(data=df, x='voluntariado', y='conexion_emocional', color='black', alpha=0.2, size=3)

plt.title('Comparaci√≥n de Conexi√≥n Emocional: Voluntarios vs No Voluntarios', fontsize=14, fontweight='bold')
plt.xlabel('¬øParticipa como Voluntario?', fontsize=12)
plt.ylabel('Conexi√≥n Emocional (1-10)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Paso 4: Prueba t de Dos Muestras
grupo_voluntarios = df[df['voluntariado'] == 'S√≠']['conexion_emocional']
grupo_no_voluntarios = df[df['voluntariado'] == 'No']['conexion_emocional']

print(f"Tama√±o grupo voluntarios: {len(grupo_voluntarios)}")
print(f"Tama√±o grupo no voluntarios: {len(grupo_no_voluntarios)}")

t_stat, p_value = stats.ttest_ind(grupo_voluntarios, grupo_no_voluntarios, alternative='greater')

print("\n" + "="*70)
print("üß™ RESULTADOS DE LA PRUEBA T DE DOS MUESTRAS\n")
print(f"Estad√≠stico t: {t_stat:.4f}")
print(f"Valor p (unilateral): {p_value:.4f}")
print(f"Nivel de significancia Œ±: 0.05")
print("="*70)
```

```python
# Paso 5: Decisi√≥n
alpha = 0.05

if p_value < alpha:
    decision = "Rechazamos H‚ÇÄ"
    conclusion = "Hay evidencia estad√≠stica de que los voluntarios tienen MAYOR conexi√≥n emocional que los no voluntarios"
else:
    decision = "No rechazamos H‚ÇÄ"
    conclusion = "No hay evidencia estad√≠stica de diferencia en conexi√≥n emocional entre grupos"

print("\nüìã DECISI√ìN ESTAD√çSTICA:\n")
print(f"Decisi√≥n: {decision}")
print(f"\nConclusi√≥n: {conclusion}")

# Tama√±o del efecto (Cohen's d)
mean_diff = grupo_voluntarios.mean() - grupo_no_voluntarios.mean()
pooled_std = np.sqrt(((len(grupo_voluntarios)-1)*grupo_voluntarios.std()**2 + 
                      (len(grupo_no_voluntarios)-1)*grupo_no_voluntarios.std()**2) / 
                     (len(grupo_voluntarios) + len(grupo_no_voluntarios) - 2))
cohens_d = mean_diff / pooled_std

print(f"\nüìè Tama√±o del efecto (Cohen's d): {cohens_d:.3f}")
print(f"Diferencia de medias: {mean_diff:.3f} puntos")
print("\nInterpretaci√≥n tama√±o del efecto: ", end="")
if abs(cohens_d) < 0.2:
    print("Efecto peque√±o (<0.2)")
elif abs(cohens_d) < 0.5:
    print("Efecto peque√±o-mediano (0.2-0.5)")
elif abs(cohens_d) < 0.8:
    print("Efecto mediano (0.5-0.8)")
else:
    print("Efecto grande (>0.8)")
```

**üìù SOLUCI√ìN 3.2b - Interpretaci√≥n:**

a) **Interpretaci√≥n en contexto:**

**Si rechazamos H‚ÇÄ:** Existe evidencia robusta de que el **voluntariado fortalece la conexi√≥n emocional** con la misi√≥n. Esto confirma la teor√≠a de Relationship Fundraising que sugiere que el involucramiento activo (no solo transaccional) genera mayor engagement. Los voluntarios no solo donan dinero, sino tiempo y esfuerzo, lo que profundiza su v√≠nculo con la organizaci√≥n.

**Implicaci√≥n estrat√©gica:** El voluntariado funciona como una "escalera de engagement" que transforma donadores transaccionales en advocates comprometidos.

b) **Magnitud del efecto:**

- **Cohen's d > 0.5:** Efecto mediano a grande, **pr√°ctico y significativo**
- **Diferencia de ~1-2 puntos en escala 1-10:** Suficientemente grande para ser notable en la experiencia real de los donadores
- **Conclusi√≥n:** El efecto no solo es estad√≠sticamente significativo, sino tambi√©n **relevante para la gesti√≥n** de la organizaci√≥n

c) **Recomendaciones estrat√©gicas:**

1. **Programa de Voluntariado Escalonado:**
   - Crear oportunidades de bajo compromiso (1 d√≠a) para donadores nuevos
   - Ofrecer roles de mayor responsabilidad para voluntarios experimentados

2. **Invitaciones Proactivas:**
   - Invitar a todos los donadores nuevos a un evento de voluntariado en sus primeros 90 d√≠as
   - Meta: 30% de donadores nuevos participen en al menos 1 actividad de voluntariado al a√±o

3. **Voluntariado Virtual:**
   - Ofrecer opciones remotas para donadores que no pueden asistir presencialmente
   - Ej: mentor√≠a virtual, revisi√≥n de documentos, gesti√≥n de redes sociales

4. **Reconocimiento:**

5. **Medici√≥n:**
   - Trackear tasa de conversi√≥n: donador ‚Üí voluntario
   - Comparar retenci√≥n entre donadores-voluntarios vs solo donadores

### 3.3 Prueba t de Dos Muestras: Retenci√≥n

```python
print("üéØ AN√ÅLISIS: Satisfacci√≥n General por Retenci√≥n\n")
print("Grupos: Retenidos (donara_proximo_a√±o = S√≠) vs No Retenidos (No)\n")
print("="*70)

# Paso 1: Hip√≥tesis
print("\nüìã HIP√ìTESIS:\n")
print("H‚ÇÄ: Œº_retenidos = Œº_no_retenidos (no hay diferencia en satisfacci√≥n)")
print("H‚ÇÅ: Œº_retenidos > Œº_no_retenidos (retenidos tienen mayor satisfacci√≥n)")
print("Tipo: Unilateral derecha")
print("Œ± = 0.05")

# Paso 2: Exploraci√≥n descriptiva
print("\n\nüìä ESTAD√çSTICAS DESCRIPTIVAS POR GRUPO:\n")
stats_retencion = df.groupby('donara_proximo_a√±o')['satisfaccion_general'].agg(['mean', 'std', 'count'])
stats_retencion.columns = ['Media', 'Desv. Est√°ndar', 'n']
print(stats_retencion)

# Paso 3: Visualizaci√≥n
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Boxplot
sns.boxplot(data=df, x='donara_proximo_a√±o', y='satisfaccion_general', palette='Set1', ax=axes[0])
sns.stripplot(data=df, x='donara_proximo_a√±o', y='satisfaccion_general', color='black', alpha=0.2, size=3, ax=axes[0])
axes[0].set_title('Boxplot: Satisfacci√≥n por Retenci√≥n', fontsize=12, fontweight='bold')
axes[0].set_xlabel('¬øDonar√° pr√≥ximo a√±o?')
axes[0].set_ylabel('Satisfacci√≥n General (1-10)')

# Violin plot
sns.violinplot(data=df, x='donara_proximo_a√±o', y='satisfaccion_general', palette='Set1', ax=axes[1])
axes[1].set_title('Violin Plot: Satisfacci√≥n por Retenci√≥n', fontsize=12, fontweight='bold')
axes[1].set_xlabel('¬øDonar√° pr√≥ximo a√±o?')
axes[1].set_ylabel('Satisfacci√≥n General (1-10)')

plt.tight_layout()
plt.show()

# Paso 4: Prueba t
grupo_retenidos = df[df['donara_proximo_a√±o'] == 'S√≠']['satisfaccion_general']
grupo_no_retenidos = df[df['donara_proximo_a√±o'] == 'No']['satisfaccion_general']

t_stat, p_value = stats.ttest_ind(grupo_retenidos, grupo_no_retenidos, alternative='greater')

print("\n" + "="*70)
print("üß™ RESULTADOS DE LA PRUEBA T:\n")
print(f"Estad√≠stico t: {t_stat:.4f}")
print(f"Valor p: {p_value:.4f}")
print(f"Œ±: 0.05")
print("="*70)

# Paso 5: Decisi√≥n
alpha = 0.05

if p_value < alpha:
    decision = "Rechazamos H‚ÇÄ"
    conclusion = "Los donadores RETENIDOS tienen significativamente mayor satisfacci√≥n general que los no retenidos"
else:
    decision = "No rechazamos H‚ÇÄ"
    conclusion = "No hay evidencia suficiente de diferencia en satisfacci√≥n"

print(f"\nüìã DECISI√ìN: {decision}")
print(f"\nCONCLUSI√ìN: {conclusion}")

# Tama√±o del efecto
mean_diff = grupo_retenidos.mean() - grupo_no_retenidos.mean()
pooled_std = np.sqrt(((len(grupo_retenidos)-1)*grupo_retenidos.std()**2 + 
                      (len(grupo_no_retenidos)-1)*grupo_no_retenidos.std()**2) / 
                     (len(grupo_retenidos) + len(grupo_no_retenidos) - 2))
cohens_d = mean_diff / pooled_std

print(f"\nüìè TAMA√ëO DEL EFECTO:")
print(f"   Diferencia de medias: {mean_diff:.3f} puntos")
print(f"   Cohen's d: {cohens_d:.3f}")
if abs(cohens_d) > 0.8:
    print(f"   Interpretaci√≥n: Efecto GRANDE (muy relevante)")
elif abs(cohens_d) > 0.5:
    print(f"   Interpretaci√≥n: Efecto MEDIANO (relevante)")
elif abs(cohens_d) > 0.2:
    print(f"   Interpretaci√≥n: Efecto PEQUE√ëO-MEDIANO")
else:
    print(f"   Interpretaci√≥n: Efecto PEQUE√ëO")
```

**üìù SOLUCI√ìN 3.3 - Interpretaci√≥n y Recomendaciones:**

### Interpretaci√≥n

**Hallazgo principal:** Los donadores retenidos tienen una satisfacci√≥n general significativamente **mayor** que los no retenidos (diferencia t√≠pica de 2-3 puntos en escala 1-10).

**Implicaciones:**

1. **Relaci√≥n causal bidireccional:**

   - Compromiso de retenci√≥n ‚Üí Mayor satisfacci√≥n (cognitive consistency)

   - Donadores con satisfacci√≥n < 6.0 tienen alto riesgo de no retenci√≥n

3. **Magnitud del efecto (Cohen's d > 1.0 t√≠picamente):**
   - Efecto **grande y pr√°ctico**
   - Una diferencia de 2-3 puntos representa un cambio sustancial en la experiencia del donador

### Recomendaciones Estrat√©gicas
