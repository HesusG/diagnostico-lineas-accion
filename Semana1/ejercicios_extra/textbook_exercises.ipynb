## üìù Ejercicio 3-84: Raiders de Rockland

**Contexto:**  
Un aficionado al f√∫tbol americano afirma: _"Los Raiders de Rockland tienen un promedio de 3.6 yardas de recorrido por tierra. Como s√≥lo necesitan 10 yardas para anotar y tienen cuatro oportunidades para lograrlo, la anotaci√≥n es segura, siempre y cuando mantengan su forma de jugar por tierra."_

**Pregunta:**  
¬øQu√© reacci√≥n tendr√≠a usted ante esta afirmaci√≥n? Analice si es v√°lido concluir que la anotaci√≥n es "segura" bas√°ndose √∫nicamente en la media.

**Pistas:**
- Considera que la media es un promedio, no una garant√≠a
- Piensa en la variabilidad de los datos
- ¬øQu√© otras estad√≠sticas necesitar√≠as conocer?

### ‚úçÔ∏è Tu respuesta (formato texto):

*Escribe aqu√≠ tu an√°lisis cr√≠tico de la afirmaci√≥n del aficionado:*

1. **Problema con la afirmaci√≥n:**
   - [Tu respuesta aqu√≠]

2. **Estad√≠sticas adicionales necesarias:**
   - [Tu respuesta aqu√≠]

3. **Conclusi√≥n:**
   - [Tu respuesta aqu√≠]

````python
# Simulaci√≥n: Genera 100 jugadas con media 3.6 pero variabilidad
# Esto demuestra por qu√© la media sola no garantiza el resultado

np.random.seed(42)
# Asumiendo distribuci√≥n normal con media 3.6 y desviaci√≥n est√°ndar 2.0
jugadas = np.random.normal(loc=3.6, scale=2.0, size=100)

print(f"Media de yardas: {jugadas.mean():.2f}")
print(f"Desviaci√≥n est√°ndar: {jugadas.std():.2f}")
print(f"Rango: [{jugadas.min():.2f}, {jugadas.max():.2f}]")

# Visualizaci√≥n
plt.hist(jugadas, bins=20, edgecolor='black', alpha=0.7)
plt.axvline(jugadas.mean(), color='red', linestyle='--', linewidth=2, label=f'Media: {jugadas.mean():.2f}')
plt.axvline(2.5, color='orange', linestyle='--', linewidth=2, label='M√≠nimo para anotar (2.5 yardas x 4)')
plt.xlabel('Yardas por jugada')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de Yardas por Jugada')
plt.legend()
plt.show()

# Probabilidad de anotar en 4 jugadas
intentos = 10000
anotaciones = 0
for _ in range(intentos):
    cuatro_jugadas = np.random.normal(loc=3.6, scale=2.0, size=4)
    if cuatro_jugadas.sum() >= 10:
        anotaciones += 1

probabilidad = anotaciones / intentos
print(f"\nüìä Probabilidad de anotar (10+ yardas en 4 jugadas): {probabilidad:.2%}")
print(f"\n‚ö†Ô∏è Conclusi√≥n: La anotaci√≥n NO es 'segura' - solo ocurre ~{probabilidad:.0%} del tiempo")
````

## üìù Ejercicio 8-64: Prueba de Hip√≥tesis Completa (Ejemplo Pr√°ctico)

**Contexto:**  
Una empresa afirma que sus empleados tienen un salario promedio de $45,000 anuales. El sindicato sospecha que el salario real es menor. Se toma una muestra aleatoria de 30 empleados y se encuentra:

- Media muestral: $42,500
- Desviaci√≥n est√°ndar muestral: $5,000
- Nivel de significancia: Œ± = 0.05

**Pregunta:**  
¬øHay evidencia suficiente para afirmar que el salario promedio es significativamente menor a $45,000?

### Paso 1: Formulaci√≥n de Hip√≥tesis

- H‚ÇÄ: # Escribe aqu√≠
- H‚ÇÅ: # Escribe aqu√≠
- **Tipo de prueba:** [Selecciona]

````python
# Paso 2: Datos del problema
mu_0 = 45000  # Media poblacional hipot√©tica
x_bar = 42500  # Media muestral
s = 5000  # Desviaci√≥n est√°ndar muestral
n = 30  # Tama√±o de muestra
alpha = 0.05  # Nivel de significancia

# Paso 3: Calcular estad√≠stico de prueba (t-test)
t_stat = # TU C√ìDIGO AQU√ç: (x_bar - mu_0) / (s / np.sqrt(n))

# Paso 4: Grados de libertad
df = n - 1

# Paso 5: Valor cr√≠tico (prueba unilateral izquierda)
t_critical = stats.t.ppf(alpha, df)

# Paso 6: P-value
p_value = stats.t.cdf(t_stat, df)

# Paso 7: Decisi√≥n
rechazar_h0 = p_value < alpha

print("=" * 70)
print("PRUEBA DE HIP√ìTESIS - EJERCICIO 8-64")
print("=" * 70)
print(f"H‚ÇÄ: Œº = ${mu_0:,}")
print(f"H‚ÇÅ: Œº < ${mu_0:,}")
print(f"\nDatos de la muestra:")
print(f"  ‚Ä¢ n = {n}")
print(f"  ‚Ä¢ xÃÑ = ${x_bar:,}")
print(f"  ‚Ä¢ s = ${s:,}")
print(f"\nEstad√≠stico de prueba:")
print(f"  ‚Ä¢ t = {t_stat:.4f}")
print(f"  ‚Ä¢ gl = {df}")
print(f"\nValor cr√≠tico (Œ± = {alpha}):")
print(f"  ‚Ä¢ t_cr√≠tico = {t_critical:.4f}")
print(f"\nP-value: {p_value:.6f}")
print(f"\n{'='*70}")
print(f"DECISI√ìN: {'RECHAZAR H‚ÇÄ' if rechazar_h0 else 'NO RECHAZAR H‚ÇÄ'}")
print(f"{'='*70}")
if rechazar_h0:
    print(f"\n‚úÖ CONCLUSI√ìN: Hay evidencia suficiente (Œ± = {alpha}) para afirmar que")
    print(f"   el salario promedio es significativamente menor a ${mu_0:,}.")
else:
    print(f"\n‚ùå CONCLUSI√ìN: No hay evidencia suficiente (Œ± = {alpha}) para afirmar que")
    print(f"   el salario promedio es menor a ${mu_0:,}.")
````

````python
# Visualizaci√≥n de la prueba de hip√≥tesis
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)

plt.figure(figsize=(12, 6))
plt.plot(x, y, 'b-', linewidth=2, label='Distribuci√≥n t (gl=29)')

# Regi√≥n cr√≠tica
x_crit = x[x <= t_critical]
plt.fill_between(x_crit, stats.t.pdf(x_crit, df), alpha=0.3, color='red', 
                 label=f'Regi√≥n de rechazo (Œ±={alpha})')

# Estad√≠stico de prueba
plt.axvline(t_stat, color='green', linestyle='--', linewidth=2.5, 
            label=f't = {t_stat:.4f}')
plt.axvline(t_critical, color='red', linestyle='--', linewidth=2, 
            label=f't_cr√≠tico = {t_critical:.4f}')

plt.xlabel('Valor t', fontsize=12)
plt.ylabel('Densidad de probabilidad', fontsize=12)
plt.title('Prueba de Hip√≥tesis: H‚ÇÅ: Œº < $45,000 (Prueba Unilateral Izquierda)', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.show()
````
